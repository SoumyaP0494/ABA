```python
import pandas as pd
import numpy as np
from scipy.stats import norm
```

```python
ap = pd.DataFrame.from_csv('MSFT.csv')
ap.drop(["PriceDiff"], axis = 1, inplace = True)
ap_data = ap.loc['2015-01-01':'2019-12-31']
ap_data.head()
```

```python
# we will use log return for average stock return of Apple

ap_data['logReturn'] = np.log(ap_data['Close'].shift(-1)) - np.log(ap_data['Close'])
```

```python
# Lets build 90% confidence interval for log return
sample_size = ap_data['logReturn'].shape[0]
sample_mean = ap_data['logReturn'].mean()
sample_std = ap_data['logReturn'].std(ddof=1) / sample_size**0.5

# left and right quantile
z_left = norm.ppf(0.1)
z_right = norm.ppf(0.9)

# upper and lower bound
interval_left = sample_mean+z_left*sample_std
interval_right = sample_mean+z_right*sample_std
print('90% confidence interval is ', (interval_left, interval_right))
```
