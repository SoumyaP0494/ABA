```python
ap_data = ap.loc['2015-01-01':'2019-12-31']
print(ap_data)
```

```python
print(ap_data.describe())
```

```python
plt.figure(figsize=(10, 8))
ap_data['Close'].plot()
plt.show()
```

```python
#now we are creating a new column Price difference & direction
```

```python
ap_data['PriceDiff'] = ap_data['Close'].shift(-1) - ap_data['Close']
```

```python
ap_data['Direction'] = [1 if ap_data['PriceDiff'].loc[ei] > 0 else 0 for ei in ap_data.index ]
```

```python
ap_data['ma200'] = ap_data['Close'].rolling(200).mean()
```

```python
plt.figure(figsize=(10, 8))
ap_data['ma200'].plot(label='MA200')
ap_data['Close'].plot(label='Close')
plt.legend()
plt.show()
```

```python
ap_data['LogReturn'] = np.log(ap_data['Close']).shift(-1) - np.log(ap_data['Close'])
```

```python
from scipy.stats import norm
mu = ap_data['LogReturn'].mean()
sigma = ap_data['LogReturn'].std(ddof=1)

density = pd.DataFrame()
density['x'] = np.arange(ap_data['LogReturn'].min()-0.01, ap_data['LogReturn'].max()+0.01, 0.001)
density['pdf'] = norm.pdf(density['x'], mu, sigma)

ap_data['LogReturn'].hist(bins=50, figsize=(15, 8))
plt.plot(density['x'], density['pdf'], color='red')
plt.show()
```

```python
prob_return1 = norm.cdf(-0.025, mu, sigma)
print('The Probability is ', prob_return1)
```
